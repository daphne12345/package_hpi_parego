```python
import os

# Define the root directory where the problems are located
ROOT_DIR = "/home/daphne/Code/hpi_parego/runs/cs_dist/Pymoo/Pymoo/MO"  # Change this to your actual path
TARGET_STRING = "HPI-ParEGO-50-quantile"
REPLACEMENT_STRING = "cs_dist"

# Walk through the directory structure
for root, dirs, files in os.walk(ROOT_DIR):
    for file in files:
        if file == "config.yaml":
            config_path = os.path.join(root, file)
            
            # Read and replace content
            with open(config_path, "r") as f:
                content = f.read()
            
            new_content = content.replace(TARGET_STRING, REPLACEMENT_STRING)
            
            # Write back only if there was a change
            if new_content != content:
                with open(config_path, "w") as f:
                    f.write(new_content)
                print(f"Updated: {config_path}")

print("Replacement complete.")
```

```python
from pathlib import Path

result_path = Path("figures/results")  # in overleaf

tex_path = Path("figures/latex")
tex_path.mkdir(exist_ok=True, parents=True)

figure_template = r"""
\begin{figure}[h]
    \centering
    \includegraphics[width=0.5\linewidth]{figures/results/critdiff/cdscenario_set_id.pdf}\hfill
    \includegraphics[width=0.5\linewidth]{figures/results/perf_over_time/performance_over_time_('scenario', 'set_id')_trials.pdf}\\
    \vspace{1mm}
    \includegraphics[width=\linewidth]{figures/results/ranking/final_per_combined_scenario_set_id.pdf}
    \caption{fig_caption}
    \label{fig_label}
\end{figure}
"""

table_template = r"""
Please find the results overview for the different scenarios and their subsets here:
\begin{table}[h]
\centering
\begin{tabular}{llr}
\toprule
Scenario & Subset & Link \\
\midrule
table_rows
\bottomrule
\end{tabular}
\end{table}
"""

table_row_template = "{scenario} & {set_id} & \\cref{{{fig_label}}}\\\\\n"

input_template = "\\input{{{latex_fn}}}\n"

complete_code = r"""
table_template

input_templates
"""

input_templates = ""
table_rows = ""
for (scenario, set_id), gdf in df.groupby(by=["scenario", "set"]):
    fig_caption = f"Scenario: {scenario}, set: {set_id}. Final Performance Overview"
    fig_label = f"fig:combinedperf_{scenario}_{set_id}"

    latex_str = figure_template
    latex_str = latex_str.replace("scenario", scenario)
    latex_str = latex_str.replace("set_id", set_id)
    latex_str = latex_str.replace("fig_caption", fig_caption)
    latex_str = latex_str.replace("fig_label", fig_label)

    latex_fn = tex_path / (fig_label.replace(":", "_") + ".tex")
    latex_fn.write_text(latex_str)

    print(latex_str)

    input_templates += input_template.format(latex_fn=result_path / "latex" / latex_fn.name[:-len(".tex")])
    table_rows += table_row_template.format(scenario=scenario, set_id=set_id, fig_label=fig_label)

complete_code = complete_code.replace("input_templates", input_templates)

table = table_template.replace("table_rows", table_rows)
complete_code = complete_code.replace("table_template", table)

description_fn = tex_path / "results.tex"
description_fn.write_text(complete_code)
print(complete_code)
```
