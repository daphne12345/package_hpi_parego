```python
import os

# Define the root directory where the problems are located
ROOT_DIR = "/home/daphne/Downloads/outputs/results/all/"  # Change this to your actual path
TARGET_STRING = "HPI-ParEGO-50-quantile"
for root_dir in os.listdir(ROOT_DIR):
    # Walk through the directory structure
    for root, dirs, files in os.walk(ROOT_DIR + root_dir):
        for file in files:
            if file == "config.yaml":
                config_path = os.path.join(root, file)
                
                # Read and replace content
                with open(config_path, "r") as f:
                    content = f.read()
                new_content = content.replace(TARGET_STRING, root_dir)

                # Write back only if there was a change
                if new_content != content:
                    with open(config_path, "w") as f:
                        f.write(new_content)
                    print(f"Updated: {config_path}")

    print("Replacement complete.")
```

```python
ROOT_DIR = "/home/daphne/Downloads/outputs/results/all/" 
for root_dir in os.listdir(ROOT_DIR):
    for root, dirs, files in os.walk(os.path.join(ROOT_DIR, root_dir)):

        if "overrides.yaml" in files:
            overrides_path = os.path.join(root, "overrides.yaml")
            
            # Extract the basedir from the overrides.yaml file
            with open(overrides_path, "r") as f:
                overrides_content = f.read()
                replacement_string = ''
                if "fanova" in overrides_content:
                    replacement_string += "fanova_"
                elif "hypershap" in overrides_content:
                    replacement_string += "hypershap_"
                elif "random" in overrides_content:
                    replacement_string += "random_"
                else:
                    replacement_string += 'SMAC'
                    
                if "adjust_cs=True" in overrides_content:
                    replacement_string += "adjust_cs_"
                if "constant=True" in overrides_content:
                    replacement_string += "constant_"
                if "constant=False" in overrides_content:
                    replacement_string += "dist_"
                    
                if "adjust_previous_cfgs=True" in overrides_content:
                    replacement_string += "adjust_prev_cfgs_"
                # if "cs_proba_hpi=True" in overrides_content:
                #     replacement_string += "_cs_proba_hpi_"
                    
                if "set_to=random" in overrides_content:
                    replacement_string += "set_to_random_"
                if "set_to=default" in overrides_content:
                    replacement_string += "set_to_default_"
                if "set_to=incumbent" in overrides_content:
                    replacement_string += "set_to_incumbent_"

                if "dynamic_decay=linear" in overrides_content:
                    replacement_string += "dynamic_decay_down"
                if "thresh=0.5" in overrides_content:
                    replacement_string += "thresh_05"
                if "thresh=0.75" in overrides_content:
                    replacement_string += "thresh_75"

            if "config.yaml" in files:
                config_path = os.path.join(root, "config.yaml")
                
                # Read and replace content in config.yaml
                with open(config_path, "r") as f:
                    content = f.readlines()
                    new_content = []
                    for line in content:
                        if line.strip().startswith("optimizer_id:"):
                            # Replace the value of optimizer_id with the directory name
                            new_content.append(f"optimizer_id: {replacement_string}\n")
                        else:
                            new_content.append(line)    
                # Write back only if there was a change
                if new_content != content:
                    with open(config_path, "w") as f:
                        f.writelines(new_content)
                    print(f"Updated: {config_path}")
```

```python
mean_perf = perf.groupby(['n_trials_norm', 'optimizer_id'])['trial_value__cost_inc_norm'].mean().reset_index()
```

```python
from numpy import trapz

# Group the data by optimizer_id and calculate the AUC for each
auc_results = {}
for optimizer_id, group in perf.groupby("optimizer_id"):
    # Sort the group by n_trials_norm to ensure proper integration
    group = group.sort_values("n_trials_norm")
    auc = trapz(group["trial_value__cost_inc_norm"], group["n_trials_norm"])
    auc_results[optimizer_id] = auc

# Sort the AUC results in descending order and get the top 10
top_10_optimizers = sorted(auc_results.items(), key=lambda x: x[1], reverse=False)[:10]
# top_10_optimizers.append(('SMAC', auc_results['SMAC']))

# Display the top 10 AUC results
for optimizer_id, auc in top_10_optimizers:
    print(f"Optimizer: {optimizer_id}, AUC: {auc}")
```

```python
import matplotlib.pyplot as plt

# Extract keys and values from the dictionary
optimizers = [item[0] for item in top_10_optimizers]
auc_values = [item[1] for item in top_10_optimizers]

# Create the bar chart with a logarithmic scale for better differentiation
plt.figure(figsize=(15, 6))
plt.barh(optimizers, auc_values, log=False)
plt.xlabel('AUC')
plt.ylabel('Optimizers')
# plt.xlim(0.1
# )  # Set x-axis limits
plt.title('Top 10 Optimizers by AUC')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest AUC at the top
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by n_trials_norm, then for each group find the optimizer(s) with the min cost
best_optimizers = mean_perf.sort_values("n_trials_norm").groupby("n_trials_norm").apply(
    lambda row: row[row['trial_value__cost_inc_norm'] == row['trial_value__cost_inc_norm'].min()]['optimizer_id'].unique()[0]
)

# Prepare to plot the results
best_optimizers_flat = []
best_costs = []
best_n_trials_norm = []

# Loop through to extract best optimizer(s) and their corresponding cost and trial norm
for n_trial_norm, optimizer in best_optimizers.items():
    # Find the minimum cost for this optimizer at the given n_trials_norm
    cost = perf[(perf['n_trials_norm'] == n_trial_norm) & (perf['optimizer_id'] == optimizer)]['trial_value__cost_inc_norm'].min()
    
    best_optimizers_flat.append(optimizer)
    best_costs.append(cost)
    best_n_trials_norm.append(n_trial_norm)

# Plot the best optimizer(s) trajectory
plt.figure(figsize=(10, 10))
plt.scatter(best_n_trials_norm, best_costs, c='blue', label="Best Optimizer(s)", alpha=0.6)

# Annotate optimizer IDs at each point
for n_trial_norm, cost, opt in zip(best_n_trials_norm, best_costs, best_optimizers_flat):
    plt.text(n_trial_norm, cost, f"{opt}", fontsize=10, verticalalignment="bottom", rotation=45)

# Plot settings
plt.xlabel("Normalized Trials")
plt.ylabel("Best Found Cost (Incumbent)")
plt.title("Best Optimizer(s) at Each Normalized Trial Step")
plt.legend()
plt.grid()
plt.show()
```

```python
from pathlib import Path

result_path = Path("figures/results")  # in overleaf

tex_path = Path("figures/latex")
tex_path.mkdir(exist_ok=True, parents=True)

figure_template = r"""
\begin{figure}[h]
    \centering
    \includegraphics[width=0.5\linewidth]{figures/results/critdiff/cdscenario_set_id.pdf}\hfill
    \includegraphics[width=0.5\linewidth]{figures/results/perf_over_time/performance_over_time_('scenario', 'set_id')_trials.pdf}\\
    \vspace{1mm}
    \includegraphics[width=\linewidth]{figures/results/ranking/final_per_combined_scenario_set_id.pdf}
    \caption{fig_caption}
    \label{fig_label}
\end{figure}
"""

table_template = r"""
Please find the results overview for the different scenarios and their subsets here:
\begin{table}[h]
\centering
\begin{tabular}{llr}
\toprule
Scenario & Subset & Link \\
\midrule
table_rows
\bottomrule
\end{tabular}
\end{table}
"""

table_row_template = "{scenario} & {set_id} & \\cref{{{fig_label}}}\\\\\n"

input_template = "\\input{{{latex_fn}}}\n"

complete_code = r"""
table_template

input_templates
"""

input_templates = ""
table_rows = ""
for (scenario, set_id), gdf in df.groupby(by=["scenario", "set"]):
    fig_caption = f"Scenario: {scenario}, set: {set_id}. Final Performance Overview"
    fig_label = f"fig:combinedperf_{scenario}_{set_id}"

    latex_str = figure_template
    latex_str = latex_str.replace("scenario", scenario)
    latex_str = latex_str.replace("set_id", set_id)
    latex_str = latex_str.replace("fig_caption", fig_caption)
    latex_str = latex_str.replace("fig_label", fig_label)

    latex_fn = tex_path / (fig_label.replace(":", "_") + ".tex")
    latex_fn.write_text(latex_str)

    print(latex_str)

    input_templates += input_template.format(latex_fn=result_path / "latex" / latex_fn.name[:-len(".tex")])
    table_rows += table_row_template.format(scenario=scenario, set_id=set_id, fig_label=fig_label)

complete_code = complete_code.replace("input_templates", input_templates)

table = table_template.replace("table_rows", table_rows)
complete_code = complete_code.replace("table_template", table)

description_fn = tex_path / "results.tex"
description_fn.write_text(complete_code)
print(complete_code)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.decomposition import PCA

# Prepare the figure layout
ncols = 2
nrows = 3
right = 0.6
h = 0.9
w = 1
factor = 30
hspace = 0.3
wspace = 0.7

fig = plt.figure(facecolor="white", figsize=(w * factor, h * factor))
gs = fig.add_gridspec(nrows=nrows, ncols=ncols, left=0.05, right=right, hspace=hspace, wspace=wspace)

# Clustered Heatmap: Performance per problem, grouped by optimizer
df_pivot = df.pivot_table(index="problem_id", columns="optimizer_id", values="trial_value__cost_inc_norm")
df_pivot = df_pivot.fillna(df_pivot.mean())  # Fill NaNs with column mean

fig_clustermap = sns.clustermap(df_pivot, cmap="viridis_r", figsize=(10, 8), annot=False, metric="euclidean", method="ward")
fig_clustermap.ax_heatmap.set_title("Final Performance per Problem (Clustered)")

# Final Performance Boxplot with Strip Plot
df_finalperf = df[["optimizer_id", "trial_value__cost_inc_norm"]].dropna()

ax1 = fig.add_subplot(gs[0, -1])
sns.boxplot(data=df_finalperf, y="optimizer_id", x="trial_value__cost_inc_norm", ax=ax1)
sns.stripplot(data=df_finalperf, y="optimizer_id", x="trial_value__cost_inc_norm", color="black", alpha=0.3, jitter=True, ax=ax1)
ax1.set_title("Final Performance (Boxplot + Strip)")

# Violin Plot for Performance Distribution
ax2 = fig.add_subplot(gs[1, -1])
sns.violinplot(data=df_finalperf, y="optimizer_id", x="trial_value__cost_inc_norm", ax=ax2, cut=0)
ax2.set_title("Final Performance (Violin Plot)")

# PCA for Optimizer Performance Clustering
ax3 = fig.add_subplot(gs[2, -1])

# Convert optimizer performance into a feature matrix
df_pca_input = df_pivot.fillna(0).T  # Transpose: Optimizers as rows
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_pca_input)
df_pca = pd.DataFrame(df_pca, columns=["PC1", "PC2"], index=df_pca_input.index)

sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue=df_pca.index, palette="tab10", ax=ax3)
ax3.set_title("Optimizer Performance Clustering (PCA)")


plt.show()
```
